```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the data from the 'data' folder
df = pd.read_csv("Hippocampal_Subfield_Volumes.csv")

# Check the data types of each column
print(df.dtypes)

# Select only numeric columns
df_numeric = df.select_dtypes(include=[np.number])

# Check if the columns are numeric
print(df_numeric.dtypes)

# Drop non-numeric columns
df = df.select_dtypes(include=[np.number])

# Check for missing data
missing_data = df.isnull().sum()
print("Missing Data:\n", missing_data)

# Handle missing data (if applicable)
df.fillna(df.mean(), inplace=True)

# Create a heatmap for correlations
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()


# Step 3: Perform basic EDA

# 3.1: Summary statistics
print(df.describe())

# 3.2: Check for missing data
missing_data = df.isnull().sum()
print("Missing Data:\n", missing_data)

# 3.3: Visualizing the distribution of numeric columns
df.hist(figsize=(10, 8))
plt.show()

# Create a heatmap for correlations with improvements
plt.figure(figsize=(14, 12))
sns.heatmap(df_numeric.corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, linecolor='white', vmin=-1, vmax=1)

# Improve aesthetics
plt.title('Correlation Heatmap of Hippocampal Subfield Volumes', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, ha='right', fontsize=12)
plt.tight_layout()  # Adjust layout to avoid overlapping labels
plt.show()

# 3.4: Visualizing correlations between numeric features
# Select only numeric columns
df_numeric = df.select_dtypes(include=[np.number])

# Handle missing data by filling with column means
df_numeric = df_numeric.fillna(df_numeric.mean())

# Create a heatmap for correlations
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.show()

# Step 4: Data Preprocessing (if needed)
# Example: Handling missing data
df.fillna(df.mean(), inplace=True)
```
